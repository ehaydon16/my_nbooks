# Exercise 3-1: Create some plots

```python
import pandas as pd
```

## Get the data

```python
mortality_data = pd.read_pickle('mortality_prepped.pkl')
mortality_data.head()
```

```python
mortality_wide = pd.read_pickle('mortality_wide.pkl')
mortality_wide.head()
```

## Visualize the data

```python
mortality_data.query('AgeGroup == "01-04 Years"').plot.scatter(x='Year', y='DeathRate')
```

```python
mortality_wide.plot()
```

```python
mortality_wide.plot.line(y=['15-19 Years'])
```

```python
mortality_wide.plot.area(legend='reverse')
```

```python
mortality_wide.query('Year in (1900,0)').plot.bar()
```

```python
##notes: the query function wouldnt work unless I had 2 years, so that is why year 0 is there
```

```python
mortality_wide.query('Year in (2000,0)').plot.bar(title='Child Mortality in 2000', xlabel='')
```

```python
mortality_data.query('Year in (2000,0)').pivot(index='AgeGroup', columns='Year', values='DeathRate').plot.barh(title='Child Mortality in Year 2000', xlabel='',legend=False)
```

```python
mortality_data.pivot(index='AgeGroup', columns='Year', values='DeathRate').plot.hist(legend=False)
```

```python
mortality_data.pivot(index='AgeGroup', columns='Year', values='DeathRate').plot.hist(bins=15)
```

```python
##notes: I don't think histograms work very well as long data...
```

```python
mortality_data.query('Year in (1900,0)').pivot(index='AgeGroup', columns='Year', values='DeathRate').plot.density(title='Mortality in 1900', legend=False, grid=True)
```

```python
mortality_data.query('Year in (1900,1925,1950,1975,2000)').groupby('AgeGroup')['DeathRate'].sum().plot.pie()
```

```python
mortality_wide.query('Year in (1900,1925,1950,1975,2000)').plot.barh(title=['Child Mortality: 01-04','Child Mortality: 05-09', 'Child Mortality: 10-14','Child Mortality: 15-19'],ylabel='Year',xlabel='Deaths', sharey=True, grid=False, legend=False, subplots=True, layout=(2,2),figsize=(10,10))
```
