```python
import pandas as pd
import seaborn as sns
```

```python
avo=pd.read_csv('avocado.csv')
avo.head()
```

```python
avo.info()
```

```python
avo['Date']=pd.to_datetime(avo['Date'])
```

```python
avo.info()
```

```python
avo1=avo.query('year==2015')
avo2=avo1.query('region=="Albany"')
avo3=avo2.query('type=="conventional"')[['Date','Total Bags','Small Bags']]
```

```python
avo3.head()
```

```python
avo3_melted=pd.melt(avo3,id_vars='Date',value_vars=['Total Bags','Small Bags'],var_name='Bags',value_name='Count')
avo3_melted
```

```python
avo3_melted.info()
```

```python
sns.relplot(date=avo3_melted, kind='line',x='Date',y='Count',hue='Bags')
##seaborn doesn't like using datetime for x it seems. I cannot find a fix in the book.
```

```python
avo.columns=avo.columns.str.replace('Total Volume','TotalVolume')
```

```python
avo['SalesVolume']=pd.qcut(avo.TotalVolume,q=4,labels=['Poor','Modest','Good','Excellent'])
avo.head()
```

```python
sns.countplot(data=avo,x='year',hue='SalesVolume')
```

```python
avo_melted=pd.melt(avo,id_vars='year',value_vars=['Small Bags','Large Bags','XLarge Bags'],var_name='Bags',value_name='Count')
avo_melted
```

```python
sns.relplot(data=avo_melted,kind='line',x='year',y='Count',hue='Bags')
```
